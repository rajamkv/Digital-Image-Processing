# *Loading libraries*

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


```

# *2. Load the images using OpenCV.*

```python
# Reading an image in OpenCV
path=r'../images/image_1.jpg'
img_1 = cv2.imread(path)

path=r'../images/image_2.jpg'
img_2 = cv2.imread(path)
path=r'../images/image_3.jpg'
img_3 = cv2.imread(path)
path=r'../images/image_4.jpg'
img_4 = cv2.imread(path)
# OpenCV reads the image in BGR format. 
# We can change that to RGB using this:
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
img_3 = cv2.cvtColor(img_3, cv2.COLOR_BGR2RGB)
img_4 = cv2.cvtColor(img_4, cv2.COLOR_BGR2RGB)
```

# *3. Display the images using matplotlib (make a 2x2 grid using subplots)*

```python
fig = plt.figure(figsize=[10,10])
ax1 = fig.add_subplot(2,2,1)
ax1.imshow(img_1)
plt.axis('off') # Remove the axes of the plot

ax2 = fig.add_subplot(2,2,2)
ax2.imshow(img_2)
plt.axis('off') # Remove the axes of the plot

ax3 = fig.add_subplot(2,2,3)
ax3.imshow(img_3)
plt.axis('off') # Remove the axes of the plot

ax4 = fig.add_subplot(2,2,4)
ax4.imshow(img_4)
plt.axis('off') # Remove the axes of the plot
plt.show()
```
