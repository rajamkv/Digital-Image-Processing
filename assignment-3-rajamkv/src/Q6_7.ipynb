```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
def rotimage(img,t,retain=0):
    t=-np.pi*t/180
    if(retain == 1):
        img2=np.zeros(img.shape)
        X=img.shape[0]//2
        Y=img.shape[1]//2
        for i in range (img.shape[0]):
            for j in range (img.shape[1]):
                i1=int((i-X)*np.cos(t)-(j-Y)*np.sin(t)+X)
                j1=int((i-X)*np.sin(t)+(j-Y)*np.cos(t)+Y)
                if (i1<0 or i1>=img.shape[0] or j1<0 or j1>=img.shape[1]):
                    continue
                img2[i][j]=img[i1][j1] 
        return img2
    if(retain == 0):
        me=max(img.shape)
        img2=np.zeros((2*me,2*me))
        img2[me//2:me//2+img.shape[0],me//2:me//2+img.shape[1]]=np.copy(img)
        img=np.copy(img2)
        img2=np.zeros(img.shape)
        X=img.shape[0]//2
        Y=img.shape[1]//2
        for i in range (img.shape[0]):
            for j in range (img.shape[1]):
                i1=int((i-X)*np.cos(t)-(j-Y)*np.sin(t)+X)
                j1=int((i-X)*np.sin(t)+(j-Y)*np.cos(t)+Y)
                if (i1<0 or i1>=img.shape[0] or j1<0 or j1>=img.shape[1]):
                    continue
                img2[i][j]=img[i1][j1] 
        return img2.astype(np.uint8)
def rotimage_rgb(img,t,retain):
    return np.stack([rotimage(img[:,:,0],t,retain),
                   rotimage(img[:,:,1],t,retain),
                   rotimage(img[:,:,2],t,retain)],axis=2)
```

```python
img=cv2.imread('../images/butterfly.jpg',0)
plt.imshow(img,'gray')
plt.show()
plt.figure(figsize=[20,20])
plt.subplot(321)
plt.imshow(rotimage(img,30,0),'gray')
plt.subplot(322)
plt.imshow(rotimage(img,45,0),'gray')
plt.subplot(323)
plt.imshow(rotimage(img,60,0),'gray')
plt.subplot(324)
plt.imshow(rotimage(img,75,0),'gray')
plt.subplot(325)
plt.imshow(rotimage(img,90,0),'gray')
plt.subplot(326)
plt.imshow(rotimage(img,180,0),'gray')
```
